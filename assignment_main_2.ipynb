```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("/content/ABC Company.xlsx -csv.csv")
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
np.random.seed(0)
df['Height']=np.random.randint(150,181,size=len(df))
df
```

```python
team_counts=df['Team'].value_counts()
team_percentages=df['Team'].value_counts(normalize=True)*100
print(f"Team Counts:\n{team_counts}")
print(f"\nTeam Percentages:\n{team_percentages}")
```

**team** **distribution**


```python
team_counts.plot(kind='bar')
plt.title("Team Distribution")
plt.xlabel("Team")
plt.ylabel("Count")
plt.show()
```

**position** **distribution**

```python
postion_counts=df['Position'].value_counts()
print(postion_counts)
```

```python
postion_counts.plot(kind='pie',autopct='%1.1f%%',colors=sns.color_palette('Set2'))
plt.title('Position Distribution')
plt.ylabel('')
plt.show()
```

**age** **group**

```python
bins = [20,25,30,35,40,45,50]
labels = ['20-25','25-30','30-35','35-40','40-45','45-50']

df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

age_group_count = df['Age Group'].value_counts()
age_group_count.plot(kind='bar')
plt.title("Age Group Distribution")
plt.show()
```

**salary** **analysis**

```python
salary_data=df.groupby(['Team','Position'])['Salary'].sum()
print(salary_data.sort_values(ascending=False))
```

```python
correlation=df[['Age','Salary']].corr()
print(correlation)

plt.scatter(df['Age'],df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary')
plt.show()
```

```python
print("Most employees team:", team_counts.idxmax())
print("Top position:", postion_counts.idxmax())
print("Dominant age group:", age_group_count.idxmax())
print("highest salary:", salary_data.idxmax())
```
